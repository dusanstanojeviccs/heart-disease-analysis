# Heart Disease Analysis

In the following notebook we will be analysing the dataset from http://archive.ics.uci.edu/ml/datasets/Heart+Disease

```python
import pandas as pd
import numpy as np

url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'

columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']

data = pd.read_csv(url, names = columns)

data = data.replace('?', np.nan)

data['ca'] = data['ca'].astype(float)
data['thal'] = data['thal'].astype(float)

data = data.dropna()
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

## Data Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

print(scaler.fit(data))
```

```python
data = pd.DataFrame(scaler.fit_transform(data), columns = columns)
```

```python
print(data.describe())
```

## Visualization

```python
import matplotlib.pyplot as plt

data.boxplot(column = columns)

plt.show()
```

```python
pd.plotting.scatter_matrix(data, figsize=(6, 6))
plt.show()
```

```python
print(data.corr())
```

```python
plt.matshow(data.corr())
plt.xticks(range(len(columns)), columns, rotation = 90)
plt.yticks(range(len(columns)), columns)
plt.title("Correlation", pad = 50)
plt.show()
```

## Model Training

```python
from sklearn.model_selection import train_test_split

X = data.drop('thal', axis=1)
Y = data['thal']

print(data.head(5))

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 5)
```

```python
from keras.layers import Dense
from keras.models import Sequential

model = Sequential()

model.add(Dense(30, input_dim = 13, activation = 'tanh'))
model.add(Dense(20, activation = 'tanh'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, Y_train, epochs = 1000, verbose = 1)
```

```python
model.evaluate(X_test, Y_test, verbose = 0)
```
